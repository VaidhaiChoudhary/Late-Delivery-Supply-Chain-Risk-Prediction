```python
import matplotlib.pyplot as plt

# Data with abbreviated algorithm names
classifiers = ['LR', 'Bagging', 'RF', 'ET', 'KNN', 'MLP', 'SVM']
f1_scores = [98.96, 99.24, 98.41, 99.24, 81.86, 98.2, 98.96]

# Create a bar chart with more visible colors
colors = ['mediumblue', 'limegreen', 'firebrick', 'dodgerblue', 'orange', 'gray', 'mediumvioletred']
plt.figure(figsize=(5, 5))
bars = plt.bar(classifiers, f1_scores, color=colors)  # Opaque bars

# Add labels and title with some adjustments for space
plt.xlabel('F1-Score')
plt.ylabel('Classification Algorithms')
plt.title('F1-Score by Classification Algorithm for DataCo')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability and right-align

# Reduce y-axis label padding for better layout
plt.tick_params(axis='y', which='both', pad=5)

# Add white text labels for scores slightly below the bar top with bottom alignment
for bar, score in zip(bars, f1_scores):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 50, f"{score:.2f}", ha='center', va='bottom', color='white', fontsize=11)  # Adjusted vertical position

# Grid lines with slightly darker color (excluding bars)
plt.grid(which='major', linestyle='--', linewidth=0.5, color='gray', alpha=0.6, zorder=0)

# Show the chart
plt.tight_layout()
plt.show()
```
